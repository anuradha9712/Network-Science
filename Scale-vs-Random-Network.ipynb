```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import math
```

# Random Network using Poisson Distribution 

```python
#random network
# use poisson distribution


avgK = 5
k = np.arange(1, 15, 0.1)

Pk = np.exp(-avgK)*np.power(avgK, k)/factorial(k)

plt.plot(k, Pk)
#plt.show()
plt.xlabel("k")
plt.ylabel("Pk")
```

 #  Poisson Distribution VS Power Law Distribution

```python
plt.plot(k, Pk, label='PoissonDist')
plt.plot(k, PowerLaw(), label='PowerLawDist')
plt.legend()
plt.xlabel("k")
plt.ylabel("Pk")
```

```python
plt.plot(np.log(k), np.log(Pk), label='PoissonDist')
plt.plot(np.log(k), np.log(PowerLaw()), label='PowerLawDist')
plt.legend()
plt.xlabel("log k")
plt.ylabel("log Pk")
```

# Distances in Random Network

```python
avgK = 5
N = np.arange(5, 10**5, 100)

d = np.log(N)/np.log(avgK)

plt.plot(N, d, label='Random Network')
plt.plot(N, np.power(N,1/3), label='3D-lattice')
plt.plot(N, np.power(N,1/2), label='2D-lattice')
#plt.plot(N, N, label='1D-lattice')

plt.legend()
plt.xlabel("N")
plt.ylabel("<d>")
```

```python
avgK = 5
N = np.arange(2, 10**5, 100)

d = np.log(N)/np.log(avgK)

plt.plot(np.log(N), np.log(d), label='Random Network')
plt.plot(np.log(N),np.log( np.power(N,1/3)), label='3D-lattice') #For a cubic lattice dmax~<d> ~ N^ (1/3).
plt.plot(np.log(N),np.log(np.power(N,1/2)), label='2D-lattice') #For a square lattice dmax~<d> ~ N^ (1/2).

plt.legend()
plt.xlabel("log N")
plt.ylabel("log <d>")
```
