# Poisson Distribution VS Power Law Distribution

```python
import numpy as np
import scipy.misc
import math
import matplotlib.pyplot as plt

# <k> = 0.12
```

```python
#for scale free network

def PowerLaw(K,gamma,Kmax):
    num=K**(-gamma);
    den=0;
    for x in range(1,Kmax+2):
        den=den+x**(-gamma)
    return num/den;

G=2.1
Pk2=[]
Kmax=13
for i in range(1,Kmax+2):
    Pk2.append(PowerLaw(i,G,Kmax))

plt.plot(K,Pk2)
plt.show()
```

```python
# for random netwoerk

def poisson(k,avg):
    e=math.exp(-avg)
    p=avg**k
    return (e*p)/scipy.special.factorial(k)

AvgK=1.2
Kmax=13
K=[x for x in range(Kmax+1)];
Pk=[]
for i in range(Kmax+1):
    Pk.append(poisson(i,AvgK));
    
    
plt.plot( K, Pk)
plt.show()
```
