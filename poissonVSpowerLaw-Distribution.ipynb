#  Poisson Distribution VS Power Law Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

```python
#random network
# use poisson distribution


avgK = 5
k = np.arange(1, 15, 0.1)

Pk = np.exp(-avgK)*np.power(avgK, k)/factorial(k)

plt.plot(k, Pk)
#plt.show()
plt.xlabel("k")
plt.ylabel("Pk")
```
